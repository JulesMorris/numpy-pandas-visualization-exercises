```python
import math
import matplotlib.pyplot as plt
```

```python
#5) 

x = range(1 , 31)

# y=0n+1 and label the curve "O(1)"
y1 = [1 for n in x]

# y=log(n) and label the curve "O(log n)"
y2 = [math.log(n) for n in x]

# y=n and label the curve "O(n)"
y3 = x

#y=nlog(n) and label it "O(n log n)"
y4 = [n * math.log(n) for n in x]

# y=n^2 and label it "O(n^2)"
y5 = [n ** 2 for n in x]

# y=2^n and label it "O(2^n)"

y6 = [2 ** n for n in x]

# y=n! and label it "O(n!)"
y7 = [math.factorial(n) for n in x]

# y=nn and label it "O(n^n)"
y8= [n ** n for n in x]

plt.figure(figsize = (16, 8))
#plt.suptitle("Jules' Big O (Notation)")
plt.title('Big O Notation', fontsize = 20)
plt.xlabel('Elements', fontsize = 12)
plt.ylabel('Operations', fontsize = 12)

plt.plot(x, y1, label = '$\mathcal{O}(1)$', c = 'indigo')
plt.plot(x, y2, label = '$\mathcal{O}(logn)$', c = 'magenta')
plt.plot(x, y3, label = '$\mathcal{O}(n)$', c = 'blueviolet')
plt.plot(x, y4, label = '$\mathcal{O}(nlogn)$', c = 'blue')
plt.plot(x, y5, label = '$\mathcal{O}(n^2$)', c = 'midnightblue')
plt.plot(x, y6, label = '$\mathcal{O}(2^m)$', c = 'fuchsia', alpha = 0.8)
plt.plot(x, y7, label = '$\mathcal{O}(n!)$', c = 'slateblue')
plt.plot(x, y8, label = '$\mathcal{O}(n^n)$', c = 'darkorchid')

plt.xlim(1, 11)
plt.ylim(0, 50)
plt.legend(loc = 'upper right', fontsize = 10)

```
