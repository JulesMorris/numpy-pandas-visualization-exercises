```python
import matplotlib.pyplot as plt
import math

# Use matplotlib to plot the following equation:
#y=x^2−x+2

#You'll need to write the code that generates the x and y points.

#Add an anotation for the point 0, 0, the origin.



x = range(-50, 50)
y= [(i ** 2 - i + 2) for i in x] #looping over the x range



plt.plot(x , y)

plt.annotate('Origin', xy =(0,0))
```

```python
#Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# 1) y=√x
# 2) y=x^3
# 3) y=2^x
# 4) y=1/(x+1)

#You can use functions from the math module to help implement some of the equations above.

x = range(0, 101) #to get even 100 range
y = [math.sqrt(n) for n in x]

plt.plot(x , y)

plt.title('$y = \sqrt{x}$') 
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
```

```python
#y=x^3
x = range(0, 11)
y = [n ** 3 for n in x]

plt.plot(x, y)

plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
#plt.xticks(range(0, 11, 2)) #add tick marks using range, start at 0, to 10, increment 2
#plt.xlim(0, 11) #sets x limits
#plt.ylim(0) #sets y limits
plt.grid(True, ls = '--') #add gridlines
```

```python
#y=2^x

x = range(0, 21)
y = [2 ** n for n in x] 

plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')

plt.xticks(range(0, 21, 2))

plt.plot(x , y)
```

```python
# y=1/(x+1)

x = range(0, 16)
y = [1/(n + 1) for n in x]

plt.title('$y = 1/(x + 1)$')
plt.xlabel('$x$')
plt.ylabel('$ 1/(x + 1)$')


plt.plot(x, y)
```

```python
# 4) Combine the figures you created in the last step into one figure where each of the 4
# equations has a different color for the points. Be sure to include a legend and an 
# appropriate title for the figure.

x = list(range(0,51))

y1 = [math.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x] 
y4 = [1/(n + 1) for n in x]

plt.plot(x, y1, c = 'r', label = '$\sqrt{x}$', alpha = 1.0)
plt.plot(x, y2, c = 'g', label = '$x^3$', alpha = 1.0)
plt.plot(x, y3, c = 'b', label = '$2^x$', alpha = 1.0)
plt.plot(x, y4, c = 'm', label = '$ 1/(x + 1)$', alpha = 1.0)

plt.ylim(0 , 10)
plt.xlim(0 , 10) 

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('Functions Graphs')
plt.legend()
```
